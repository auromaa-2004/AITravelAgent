```python
import json
from functools import wraps
import requests

def tool(fn):
    """
    Custom decorator to turn a function into a tool.
    This stores metadata about the function for an AI agent to use.
    """
    @wraps(fn)
    def wrapper(*args, **kwargs):
        return fn(*args, **kwargs)  # Call the original function

    # Add metadata so the AI agent can understand the tool
    wrapper.name = fn.__name__  # Function name
    wrapper.description = fn.__doc__  # Function docstring (description)

    return wrapper  # Return the modified function
```

## **WEATHER**

```python
from google.colab import userdata
weather_api=userdata.get('weather_api_key')
```

```python
import requests
import os
import requests
from datetime import datetime, timedelta

def get_future_weather(city, future_date):
    api_key =weather_api
    url = f"https://api.openweathermap.org/data/2.5/forecast?q={city}&appid={api_key}&units=metric"

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()

        # Convert future_date to timestamp
        future_timestamp = datetime.strptime(future_date, "%Y-%m-%d").timestamp()

        closest_forecast = None
        min_time_diff = float('inf')

        # Find the closest forecast time to the requested date
        for forecast in data["list"]:
            forecast_timestamp = forecast["dt"]
            time_diff = abs(forecast_timestamp - future_timestamp)

            if time_diff < min_time_diff:
                min_time_diff = time_diff
                closest_forecast = forecast

        if closest_forecast:
            weather = {
                "Date": future_date,
                "City": data["city"]["name"],
                "Temperature": closest_forecast["main"]["temp"],
                "Weather": closest_forecast["weather"][0]["description"],
                "Humidity": closest_forecast["main"]["humidity"],
                "Wind Speed": closest_forecast["wind"]["speed"]
            }
            return weather
        else:
            return "No forecast available for this date."
    else:
        return f"Error: {response.status_code}, {response.text}"
```
